```python
from abc import ABC, abstractmethod
# ABC - Abstract Base Class
```

#### Класс унаследован от ABC, без abstractmethod

```python
class Polygon(ABC): 
  
    def noofsides(self):
        print("I have many sides")
        
```

```python
# Polygon унаследован от АВС, но всё равно можно сделать его инстанс и юзать его методы
P = Polygon()
P.noofsides()
```

```python
# классы унаследованы от Polygon, метод noofsides переопределен
class Triangle(Polygon): 
  
    def noofsides(self): 
        print("I have 3 sides") 
  
class Pentagon(Polygon): 
  
    def noofsides(self): 
        print("I have 5 sides") 
```

```python
T = Triangle() 
T.noofsides() 
  
Pe = Pentagon() 
Pe.noofsides() 
```

```python
class Hexagon(Polygon): 
  
    def no_of_sides(self): # не переобределили метод noofsides, написали свой
        print("I have 6 sides")  
```

```python
He = Hexagon() # всё работает, никаких ошибок
He.no_of_sides() 
```

```python
He.noofsides() # идет в суперкласс и ищет метод там
```
