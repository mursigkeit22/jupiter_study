```python

def brute_my(num:int, ite):
    def rec (num, ite, elem):
        num = num-1
        if num <= 0:
            print(elem)
            return
        for el in ite:
            temp = str(elem)+str(el)
            rec(num, ite, temp)
        return elem
    for el in ite:
        rec(num,ite, el)
brute_my(3, [[1,2],[3,4]])
```

Функция из видео:

```python

def brute_v(n,m):
    if type(m)==int:
        m = range(1,m+1)
    printlist = [None]*n
    def rec(idx):
        if idx == n:
            print(printlist)
            return
        for el in m:
            printlist[idx] = el
            rec(idx+1)
    rec(0)

brute_v(2, '10')   
    
```
