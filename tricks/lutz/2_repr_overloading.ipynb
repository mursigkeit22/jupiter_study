```python
class Person:
    def __init__(self, name):
        self.name = name
        
    def __repr__(self):
        return f"Instance of Person class with name {self.name}"
```

```python
cat = Person('Kat')
cat
```

```python
print(cat)
```

```python
class Person_str:
    def __init__(self, name):
        self.name = name
    def __str__(self):
        return f"Instance of Person_str class with name {self.name}"
```

```python
tom = Person_str("Tom")
tom
```

```python
print(tom)
```
