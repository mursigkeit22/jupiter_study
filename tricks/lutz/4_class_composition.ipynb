```python
class Person:
    def __init__(self, name, job = None, pay = 0):
        self.name = name
        self.job = job
        self.pay = pay
    
    def last_name(self):
        return self.name.split()[-1]
    
    def give_raise(self, percent_dec):
        self.pay = int(self.pay * (1 + percent_dec))
        
    def __repr__(self):
        return f"Person: {self.name}, {self.job}, {self.pay}"
```

```python
class Manager(Person):
    def __init__(self, name, pay):
        Person.__init__(self, name, 'manager', pay)
    
    def give_raise(self, percent_dec, bonus=.10):
        Person.give_raise(self, percent_dec + bonus)
        
    
```

```python
cat = Person('Kat', pay = 10000)
print(cat)
cat.give_raise(0.2)
print(cat)
tom = Manager('Tom Jones', 15000)
print(tom)
print(tom.last_name())
tom.give_raise(0.15)
print(tom)
```

```python
class Department:
    
    def __init__(self, *args: Person):
        self.members = list(args)
    def addMember(self, person):
        self.members.append(person)
    def give_raises(self, percent_dec):
        for person in self.members:
            person.give_raise(percent_dec)
    def show_all(self):
        for person in self.members:
            print(person)
    
    
```

```python
command = Department(cat, tom)
```

```python
command.show_all()
```
