```python
class Person:
    def __init__(self, name, job = None, pay = 0):
        self.name = name
        self.job = job
        self.pay = pay
    
    def last_name(self):
        return self.name.split()[-1]
    
    def give_raise(self, percent_dec):
        self.pay = int(self.pay * (1 + percent_dec))
        
    def __repr__(self):
        return f"Person: {self.name}, {self.job}, {self.pay}"
```

```python
class Manager:
    def __init__(self, name, pay):
        self.person = Person(name, 'manager', pay)
    
    def give_raise(self, percent_dec, bonus=.10):
        self.person.give_raise(percent_dec + bonus)
        
    def __getattr__(self, attr):
        return getattr(self.person, attr)
    
    def __repr__(self): # must be redefined
        return str(self.person)
```

```python
cat = Person('Kat', pay = 10000)
cat
```

```python
cat.give_raise(0.2)
cat
```
