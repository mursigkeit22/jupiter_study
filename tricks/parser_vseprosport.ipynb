```python
import requests
import json
from bs4 import BeautifulSoup
```

```python
url = "https://www.vseprosport.ru/reyting-bukmekerov/"
basic_url = "https://www.vseprosport.ru"
session = requests.Session()
response = session.get(url)
response
```

```python
print(response.headers)
```

```python
page = BeautifulSoup(response.content, 'html.parser')
table = page.find('div', class_='bookmeker_table')
bookmakers_urls = []
for el in table.findAll('a', {"target" : False}, class_='a_flex', ):
    temp_url = basic_url+el['href']
    bookmakers_urls.append(temp_url)
print(bookmakers_urls)
```
