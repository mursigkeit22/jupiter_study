#### Счетчик, сколько раз вызвали функцию


```python
class FunctionHolder1(object):
    def __init__(self, function):
        self.func = function
        self.called_count = 0
    def __call__(self, *args, **kwargs):
        try:
            return self.func(*args, **kwargs)
        finally:
            self.called_count += 1
```

```python
def held(function): # эту лишнюю сущность делать вообще необязательно
    return FunctionHolder1(function)

@held
def i_am_counted():
    pass

@held
def i_am_nuts():
    pass

@FunctionHolder1
def i_am_effective(val):
    return val

@i_am_effective
def just_a_func():
    pass

i_am_counted()
i_am_counted()
i_am_counted()
i_am_nuts()
i_am_effective('a ')
i_am_effective('a ')
just_a_func()

print(i_am_counted.called_count)
print(i_am_nuts.called_count)
print(i_am_effective.called_count)

```

```python
class FunctionHolder2(object):
    def __init__(self, function):
        self.func = function
        self.called_count = 0
    def __call__(self, *args, **kwargs):
        self.called_count += 1
        return self.func(*args, **kwargs)

            
def held(function):
    return FunctionHolder2(function)

@held
def i_am_counted():
    pass    
```

## Декоратор класса

Decorators can also be applied to class definitions:



```python
import re

def constructor(type_): # передаем тайп на вход
    def decorator(method): # декоратор применяется к методу
        method.constructs_type = type_ # у метода появляется атрибут констрактор тайп, значение которого тайп
        return method # возвращаем метод с добавленным атрибутом
    return decorator

def register_constructors(cls): # декоратор класса (получает класс на вход, т.е. записывается сверху класса с собакой)
    for item_name in cls.__dict__: #Выдает dictionary,в котором ключом является имя_атрибута, значением - значение_атрибута
        #(см. Пайчарм), типа {'__module__': '__main__', '__init__': <function A.__init__ at 0x017EB618>, 'foo': <function A.foo at 0x037CFB70>,
        item = getattr(cls, item_name) # значение атрибута (типа __main__, <function A.__init__ at 0x017EB618>, <function A.foo at 0x037CFB70>)
        if hasattr(item, 'constructs_type'): # если у item, т.е.у значения атрибута, есть атрибут constructs_type, 
            # то добавляем в словарь (у декорируемого класса должен быть словарь constructors, без него ошибка):
            cls.constructors[item.constructs_type] = item 
            # ключ в словаре - это тип, который мы записали с помощью декоратора constructor,
            # значение - в нашем случае метод класса, функция, надо которой мы написали @constructor(type)
    return cls # возвращаем класс
```

The constructor decorator annotates each method, the register_constructors class decorator registers each of these on the class, and the from_auto method uses this information to dispatch on the input type.

```python
@register_constructors
class IntStore(object):
    constructors = {}
    
    def __init__(self, value):
        self.value = value
        
    @classmethod # значит,работает с классом, не только с экземпляром класса
    @constructor(int) # передаем тайп инт на вход
    def from_int(cls, x): # на вход класс и некое вэлью
        return cls(x) # возвращаем класс от вэлью
    
#     def constructor(type_): # передаем тайп на вход
#     def decorator(method): # декоратор применяется к методу
#         method.constructs_type = type_ # у метода появляется атрибут констрактор тайп, который тайп
#         return method # возвращаем улучшенный метод
#     return decorator
    
    
    @classmethod
    @constructor(float)
    def from_float(cls, x):
        return cls(int(x)) # переделываем флоат в инт
    
    @classmethod
    @constructor(str)
    def from_string(cls, x):
        match = re.search(r'\d+', x) # ищет первые цифры
        if match is None:
            return cls(0) # если нет цифр, то возвращаем класс от нуля. Косяк, я считаю. по какой причине отсутствие цифр приравнивается к нулю?
        return cls(int(match.group())) # возвращаем интеджер от найденных цифр
    
    @classmethod # тоже можно применять к классу и к экземпляру класса
    def from_auto(cls, x):
        constructor = cls.constructors[type(x)] # отлавливаем тип переданного элемента, находим,
 #какой метод в нашем словаре конструкторов ему соответствует, записываем его в переменную constructor
        return constructor(x) # и возвращаем этот метод от этого икса, т е по факту результат применения функции
```

 ## Red and Blue 
    Decorators for annotation

Decorators can add annotations to functions when they're declared. For instance, suppose we want to label two types of functions, "red" functions and "blue" functions:

```python
def red(fn):
    fn.color = 'red'
    return fn

def blue(fn):
    fn.color = 'blue'
    return fn

@blue
def combine(a, b):
    result = []
    result.extend(a)
    result.extend(b)
    return result

@red
def unsafe_combine(a, b):
    a.extend(b)
    return a

@blue
def sane_combine1(a,b):
    return a+b

@blue
def sane_combine2(a, b):
    return [*a, *b]


@blue
def combine_and_save(a, b):
    result = a + b
    with open('combined', 'w') as f:
        f.write(str(result))
    return result
```

Юзается это так:

```python
def combine_using(fn, a, b):
    if hasattr(fn, 'color') and fn.color == 'red':
        print("Sorry, only blue functions allowed here!")
        return sane_combine1(a, b)  # fall back to default implementation
    return fn(a, b)

a = [1, 2]
b = [3, 4]
print(combine_using(unsafe_combine, a, b))
a
```

the red and blue decorators offer the benefits of being:

- highly and immediately visible

- inherently closely attached to/above the function definition

- consistent and foolproof (no room for typos).

## Decorators for registration


Sometimes, we want to have a centralized means of discovering a number of different functions. Decorators are a convenient means of doing this.

```python
FUNCTION_REGISTRY = []

def registered(fn):
    FUNCTION_REGISTRY.append(fn)
    return fn

@registered
def step_1():
    print("Hello")
    
@registered
def step_2():
    print("world!")
```

```python
def run_all():
    for function in FUNCTION_REGISTRY:
        function()
        
run_all()
```

```python
[step_1(), step_2()]
```

```python
(step_1(), step_2())
```

```python
step_1(), step_2()
```

Again, while we could have accomplished the same thing by doing:



```python
def step_1():
    print("Hello")
    
def step_2():
    print("world!")
    
FUNCTION_REGISTRY = [step_1, step_2]

def run_all():
    for function in FUNCTION_REGISTRY:
        function()
        
run_all()
```

```python
for el in FUNCTION_REGISTRY:
    el()
```

This means that, looking at the definition of step_1 in isolation, we don't know that it's been registered and, further, if we want to figure out how or why it's being run, we need to first identify that it's included in FUNCTION_REGISTRY, and then see that FUNCTION_REGISTRY is used in run_all. Additionally, if we add a step_3 - possibly referencing our other step functions - we have to remember to add it to FUNCTION_REGISTRY when we're done. This is harder to forget when you see step_1 and step_2 very visibly decorated with @registered.

## Unbalanced Brackets

```python
def brackets_balanced(s):
    brackets = {
        opening: closing
        for opening, closing in 
        '() {} []'.split()
    }
    closing = set(brackets.values())
    stack = []
    for char in s:
        if char not in closing:
            if char in brackets:
                stack.append(brackets[char])
            continue
        try:
            expected = stack.pop()
        except IndexError:
            return False
        if char != expected:
            return False
    return not stack
```

```python
def ensure_brackets_balanced(fn):
    for const in fn.__code__.co_consts:
        if not isinstance(const, str) or brackets_balanced(const):
            continue
        print(
            "WARNING - {.__name__} contains unbalanced brackets: {}".format(
                fn, const
            )
        )
    return fn
```

```python
@ensure_brackets_balanced
def get_root_div_paragraphs(xml_element):
    return xml_element.xpath("//div[not(ancestor::div]/p")
```
