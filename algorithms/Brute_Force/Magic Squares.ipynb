```python
def summa(n):
    return sum(range(n**2+1))//n
```

```python
def initial(n):
    m = n**2
    numbers = list(range(1, m+1))
    S = summa(n)
    all_vars = []
    good_vars=[]
    empty_list = [None]*m
    print(m,numbers,S, all_vars, good_vars, empty_list)
    return n, m,numbers,S, all_vars, good_vars, empty_list
```

```python
#делаем из списка квадрат (список списков)
def change(num_list, n):
    square = []
    temp = []
    for el in num_list:
        if len(temp) != n:
            temp.append(el)
        if len(temp)==n:
            square.append(temp)
            temp = []
    return square
```

```python
#проверяем квадрат, на вход - список
def check(line):
    diag_left = []
    diag_right = []
        
    square = change(line, n) # делаем из списка квадрат
    for i in range(n):
        if sum([el[i] for el in square]) != S:
            return False
        diag_left.append(square[i][i]) 
        diag_right.append(square[i][n-1-i])
    if sum(diag_left) != S or sum(diag_right) != S:
            return False

        
    return True

#проверяем горизонтали квадрата
def checkline(line):
    last_three = line[-1:-(n+1):-1]
    
    if sum(last_three) != S:
        return False
    return True
     
    
```

```python


def rec(i, temp, numm):
#     print('in rec, i =', i)
    if i != 0 and i % n == 0:
        
        short = temp[0:i]
#         print(i, short)
        if not checkline(short):
            return
            
        
    if i>m-1:
#         print('i>m-1')
        all_vars.append('hop')
        if check(temp):
            good_vars.append('hop')
            print(temp)
        return
    for el in numm:
        temp[i]=el
        temp_numm = numm[:]
        temp_numm.remove(el)
        rec(i+1, temp, temp_numm)


n, m,numbers,S, all_vars, good_vars, empty_list = initial(4)
rec(0, empty_list, numbers)
print(len(all_vars))
print(len(good_vars))

```
