```python
# сделать визуальное представление
def heapify_min(array, ind, maxind):
    left = ind*2+1
    right = ind*2+2
    smallest = ind
    if left < maxind and array[left] < array[smallest]:
        smallest = left
    if right < maxind and array[right] < array[smallest]:
        smallest = right
    if smallest != ind:
        array[ind], array[smallest] = array[smallest], array[ind]
        heapify_min(array, smallest, maxind)
    return array
```

```python
def buildHeap(array):
    A = array[0:]# чтобы не менять оригинальный аррей, его надо присваивать другой переменной (срез или дипкопи)
    n = len(A)
    for i in range (n//2-1, -1, -1):
        heapify_min(A, i, n)
        
    return A
```

```python
def largestK(array, k):
    A_heap = buildHeap(array[:k])
    for el in A[k:]:
        if el > A_heap[0]:
            A_heap[0] = el
            heapify_min(A_heap, 0, k)
    return A_heap[0]
        
    
```
