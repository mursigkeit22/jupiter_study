Find k-th smallest element in collection

(the largest one of five smallest)

```python
def heapify(array, ind, maxind):
    left = ind*2+1
    right = ind*2+2
    largest = ind
    if left < maxind and array[left] > array[largest]:
        largest = left
    if right < maxind and array[right] > array[largest]:
        largest = right
    if largest != ind:
        array[ind], array[largest] = array[largest], array[ind]
        heapify(array, largest, maxind)
    return array
        
```

```python
def buildHeap(A): # если не хочу менять оригинальный аррей, его надо присваивать другой переменной
    n = len(A)
    for i in range (n//2-1, -1, -1):
        heapify(A, i, n)
        
    return A
```

```python
def smallest_k(array, k):
    k_array = array[0:k]
    remain = array[k:]
    k_array = buildHeap(k_array)
    for el in remain:
        if el < k_array[0]:
            k_array[0] = el
            heapify(k_array, 0, k)
    return k_array[0]
    

smallest_k(B, 6)
```
