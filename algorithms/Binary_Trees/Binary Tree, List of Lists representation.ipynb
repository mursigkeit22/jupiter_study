```python
def binary_tree(value):
    return [value, [], []]
```

```python
def insert_left(root, new_left): # where root is binary tree, new_left is a single int or string
    old_left = root.pop(1)
    if len(old_left) > 0: # 0 or 1???
        root.insert(1, [new_left, old_left, []])
    else:
        root.insert(1, [new_left, [], []])
    return root
```

```python
def insert_right(root, new_right): # where root is binary tree, new_left is a single int or string
    old_right = root.pop(2)
    if len(old_right) > 0: # 0 or 1???
        root.insert(2, [new_right, [], old_right])
    else:
        root.insert(2, [new_right, [], []])
    return root
```

```python
def get_root_val(root):
    return root[0]
def set_root_val(root, new_val):
    root[0] = new_val
def get_left_child(root):
    return root[1]
def get_right_child(root):
    return root[2]
```

```python
# [0, [3, [], []], [2, [], []]]

young_tree = binary_tree(0)
insert_right(young_tree, 8)
insert_right(young_tree, 6)
insert_right(young_tree, 3)

insert_left(young_tree, 5)
insert_left(young_tree, 4)
insert_left(young_tree, 2)
insert_left(young_tree[2][2], '*')

# insert_right(young_tree[2], 1)

# insert_right(young_tree[1], 4)
print(young_tree)
```
