## Изменяемые\неизменяемые значения

```python
def extender(source_list, extend_list):
    source_list.extend(extend_list)
values = [1,2,3]
print('id(values)', id(values))
extender(values, [4,5,6])
print(values)
print('id(values)', id(values))
```

```python
def extender(source_list, extend_list):
    source_list.extend(extend_list)
    return source_list
values = [1,2,3]
print(extender(values, [4,5,6]))
print(values)
```

```python
def replacer(source_tuple, replace_with):
    source_tuple = replace_with
    return source_tuple
print('id(replacer)', id(replacer))
user_info = ('Anya', 25)
print('id(user_info)', id(user_info))
print(replacer(user_info, ('Ola', 26)))
print('id(replacer)', id(replacer))

print(user_info)
print('id(user_info)', id(user_info))
```

```python
def replacer(source_tuple, replace_with):
    source_tuple = replace_with

user_info = ('Anya', 25)
print(replacer(user_info, ('Ola', 26)))
print(user_info)
```

```python
def replace_tuple(tuplee = ()):
    temp_list = []
    for el in tuplee:
        temp_list.append(el+1)
    tuplee = tuple(temp_list)
    return tuplee
replace_tuple((1,))
```

### Звездочки

```python
def printer(*args):
    print(type(args))
    for argument in args:
        print(argument)
printer(1,2,3,4,5)
```

```python
name_list = ['John', 'Bill', 'Amy']
printer(*name_list)
```

```python
def summ_plus_tuple(*args: tuple) -> tuple:
    summ = (0,)
    for el in args:
        el = el+(1,)
        summ+=el
    return summ

summ_plus_tuple((1,2,3), (8,9))
```

```python
def summ_plus_one(*args) -> int:
    summ = 0
    for el in args:
        el = el+1
        summ+=el
    return summ

summ_plus_one(*(1,2,3), *(8,9))
```

### kwargs

```python
def printer(**kwargs):
    print(type(kwargs))
    print(kwargs)
    for key, value in kwargs.items():
        print('{}: {}'.format(key, value))
printer(a=10, b = 11)
```

```python
printer(mur = 11, pur = 12)
```

```python
my_dict = {'mur': 11, 'pur': {'shkur':12, 'ururur':13}}
printer(**my_dict)
```

### Функциональное программирование

```python
def caller (func, params):
    return func(*params)
```

```python
def my_func(*args):
    summa = 0
    for arg in args:
        summa += sum(arg)
    return summa
my_func([5,6])
```

```python
caller(my_func, [[1,2],[2,3], [9,8]])
```

```python
def get_multiplier():
    def inner(a,b):
        return a*b
    return inner
multiplier = get_multiplier()
multiplier(2,3)
```

```python
def get_multiplier(number):
    def inner(a):
        return a*number
    return inner
multiplier = get_multiplier(2)
multiplier(10)
```

### partial

```python
from functools import partial
```

```python
def greeter(person, greeting):
    return '{}, {}!'.format(greeting, person)
```

```python
hier = partial(greeter, greeting='hi')
helloer = partial(greeter, greeting = 'hello')
print(hier('brother'))
print(helloer('sir'))
```
