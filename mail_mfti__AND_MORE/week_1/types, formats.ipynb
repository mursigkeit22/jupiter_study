```python
class ClassList():
    def __init__(self,name, value):
        self.name = name
        self.value = value
class_obj_list = (ClassList('a',None), ClassList('b',''), ClassList('c',0), ClassList('d', []))
for el in class_obj_list:
    print(el.name, el.value)
```
